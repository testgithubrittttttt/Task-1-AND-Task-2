```python
import torch
from PIL import Image, ImageEnhance
from diffusers import StableDiffusionPipeline
import os

# Get user input for the prompt
def get_user_input():
    prompt = input("Enter a description for the scene (e.g., 'A cozy kitchen with natural light'): ")
    if not prompt.strip():
        prompt = "A cozy kitchen with natural light coming through a window."
    return prompt

# Dynamically resize the object based on background size
def dynamic_resize(object_image, background):
    bg_width, bg_height = background.size
    # Adjust the size to fit the coffee cup better in the scene
    object_width = int(bg_width * 0.15)  # 15% of the background width
    object_height = int(object_image.height * (object_width / object_image.width))  # Maintain aspect ratio
    object_image = object_image.resize((object_width, object_height), Image.LANCZOS)
    return object_image

# Place the object dynamically based on the scene (e.g., on the countertop)
def place_object(background, object_image, position="counter"):
    bg_width, bg_height = background.size
    obj_width, obj_height = object_image.size

    # Place near the countertop area (centered horizontally and placed higher)
    x = (bg_width - obj_width) // 2
    y = bg_height - obj_height - 300  # Adjust the Y-axis to place the cup higher on the counter

    # Paste the coffee cup onto the background using the alpha channel for transparency
    background.paste(object_image, (x, y), object_image)
    return background

# Generate a background based on the text prompt
def generate_background(prompt, device):
    # Load the Stable Diffusion pipeline for text-to-image generation
    pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5").to(device)

    # Generate the background image using the provided prompt
    background = pipe(prompt, guidance_scale=7.5).images[0]
    return background

# Main function to generate the final image
def generate_image(image_path, output_path):
    device = "cuda" if torch.cuda.is_available() else "cpu"
    prompt = get_user_input()

    # Load and process the coffee cup image (input object image)
    try:
        object_image = Image.open(image_path).convert("RGBA")
    except Exception as e:
        print(f"Error loading image: {e}")
        return

    # Generate the background based on the user's prompt
    background = generate_background(prompt, device)

    # Dynamically resize and place the object (coffee cup) based on the background
    object_image = dynamic_resize(object_image, background)
    final_image = place_object(background, object_image)

    # Save the final generated image
    final_image.save(output_path)
    print(f"Image saved to {output_path}")

# Upload the coffee cup image and run the code
from google.colab import files
uploaded = files.upload()

# Get the uploaded image's path and set the output file path
image_path = list(uploaded.keys())[0]
output_path = './generated_output.png'

# Generate the image with the coffee cup composited onto the background
generate_image(image_path, output_path)

# Display the generated image
from IPython.display import Image as IPImage, display
display(IPImage(output_path))
```
